```python
df = pd.read_csv("../input/csic-2010-web-application-attacks/csic_database.csv")
df.head(10)
```

```python
df.info()
```

```python
df2 = df[["Method", "host", "cookie", "lenght", "content", "URL","classification"]] 
df2.head(10)
```

```python
df2.rename(columns={ df2.columns[3]: "length" }, inplace=True)
```

```python
df2
```

```python
#clean"cookie"
df2 = pd.merge(df2, pd.DataFrame(df2['cookie'].str.split('=',expand=True)), how='left', left_index=True, right_index=True)
df2.rename(columns={ df2.columns[7]: "useless1" }, inplace=True)
df2.rename(columns={ df2.columns[8]: "cookie_JSESSIONID" }, inplace=True)
```

```python
df2
```

```python
#clean"length"
df2 = pd.merge(df2, pd.DataFrame(df2['length'].str.split(':',expand=True)), how='left', left_index=True, right_index=True)
df2.rename(columns={ df2.columns[9]: "useless2" }, inplace=True)
df2.rename(columns={ df2.columns[10]: "content_length" }, inplace=True)
```

```python
#clean"URL"
df3 = df2
df3 = pd.merge(df3, pd.DataFrame(df3['URL'].str.split(' ',expand=True)), how='left', left_index=True, right_index=True)
df3.rename(columns={ df3.columns[11]: "clean_URL" }, inplace=True)
df3.rename(columns={ df3.columns[12]: "HTTP_ver" }, inplace=True)
```

```python
#try split URL
from urllib import parse as ps
my_url = ps.urlparse('http://netloc/path;parameters?query=argument#fragment')
print(my_url)
#scheme 是协议  netloc 是域名服务器  path 相对路径  params是参数 query是查询的条件
print(my_url.username)
print(my_url.password)
print(my_url.hostname)  #netloc in lower case
print(my_url.port)
#username と password は入力 URL に含まれるときに利用できて、含まれないときは None です。 hostname は netloc と同じ値で全て小文字です。 port は存在するときに整数値へ変換され、存在しないときは None です。
#https://learnku.com/docs/pymotw/urllibparse-split-urls-into-components/3432
```

```python
#split the URL
df3['URL_scheme'],df3['URL_netloc'],df3['URL_path'],df3['URL_params'],df3['URL_query'], df3['URL_fragment'] = zip(*df3['clean_URL'].map(ps.urlparse))
```

```python
df4 = df3
df4 = df3[["Method", "host", "cookie_JSESSIONID", "content", "content_length", "clean_URL", "HTTP_ver", "URL_scheme", "URL_netloc", "URL_path", "URL_params", "URL_query", "URL_fragment", "classification"]]
df4.head(20)
```
