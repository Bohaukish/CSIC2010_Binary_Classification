```python
def support(inpath, outpath):
    df = pd.read_csv(inpath)  #read
    df_only1 = df[df['label'] == 1]  # filter1: 1-labelled entries
    
    # filter: remove two certain ATK types quickly 不要用drop好他妈慢
    # #df_only1 = df_only1.drop(df_only1[(df_only1['attack_type'] == 'sensitive_file_detection') | (df_only1['attack_type'] =="suspicious_file_access")].index) #删除x小于0.01或大于10的行
    y = df_only1[df_only1["attack_type"].str.contains('_file_')]
    list_wanna_delete = list(y.attack_type)
    list_ori = list(df_only1.attack_type)
    ret = list(set(list_wanna_delete) ^ set(list_ori))  #求差集
    result = df_only1[df_only1.attack_type.isin(ret)]

    df_see = result.groupby(["IP", "time"]).size().reset_index(name='Freq') # filter2: unique [ip, time] combined attributes
    df_multi = df_see[df_see['Freq']> 1]  #filter3: freq > 1
    df_multi.to_csv(outpath)   # save
```

```python
for i in range(1, 19):
    inpath = "/kaggle/input/hikari/v3_" + str(i) + "_labelled.csv"
    outpath = "/kaggle/working/support_" + str(i) + ".csv"
    support(inpath, outpath)
```
